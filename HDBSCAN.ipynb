```python
import pandas as pd
import numpy as np
import hdbscan
import sklearn
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('glist_2.csv')
```

```python
X, y = [data[['RAJ2000_gal', 'DEJ2000_gal', 'z_gal', 'RAJ2000_group', 'DEJ2000_group', 'z_group']], data['iGrID']]
```

```python
clust = hdbscan.HDBSCAN(min_cluster_size=30)
```

```python
pred = clust.fit_predict(X)
```

```python
sklearn.metrics.v_measure_score(pred,y)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X.RAJ2000_gal, X.DEJ2000_gal, c=pred)
plt.show()
```
