```python
import pandas as pd
```

```python
grlist_2, galist_2  = [pd.read_csv("grlist_2.csv"), pd.read_csv("galist_2.csv")]
```

```python
grlist_s, galist_s = [pd.read_csv("grlist_s.csv"), pd.read_csv("galist_s.csv")]
```

* grlist_2	-  List of groups and their parameters detected in 2MRS (Table A.1) (31506 rows)
* grlist_s	-  List of groups and their parameters detected in SDSS DR12 (Table A.2) (229893 rows)
* galist_2	-  List of 2MRS galaxies used, as well as their IDs and coordinates (Table A.6) (43425 rows)
* galist_s	-  List of SDSS galaxies used, as well as their IDs and coordinates (Table A.7) (402439 rows)

```python
glist_2 = galist_2.merge(grlist_2, left_on = 'iGrID', right_on= 'iGrID', how = 'left')
glist_2 = glist_2[glist_2['Ntot'] > 30]
```

```python
glist_s  = galist_s.merge(grlist_s, left_on = 'iGrID', right_on= 'iGrID', how = 'left')
glist_s  = glist_s[glist_s['Ntot'] > 30]
```
